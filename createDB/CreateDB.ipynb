```python
import json
import string
import time
import os
```

```python
start_time = time.time()
Mturk_feminist_usernames = []
Mturk_misogynist_usernames = []
Network_feminist_usernames = []
Network_misogynist_usernames = []
Nina_feminist_usernames = []
Nina_msiogynist_usernames = []
stanford_feminist_usernames = []
stanford_msiogynist_usernames = []


with open("feminist_usernames","r") as Mturk_feminist_file:
    for line in Mturk_feminist_file.readlines():
        Mturk_feminist_usernames.append(line.strip("\n"))

with open("misogynist_usernames","r") as Mturk_misogynist_file:
    for line in Mturk_misogynist_file.readlines():
        Mturk_misogynist_usernames.append(line.strip("\n"))

# Louvain method
with open("network_feminist","r") as Network_feminist_file:
    for line in Network_feminist_file.readlines():
        Network_feminist_usernames.append(line.strip("\n"))

with open("network_misogynist","r") as Network_misogynist_file:
    for line in Network_misogynist_file.readlines():
        Network_misogynist_usernames.append(line.strip("\n"))

# This is the output of CLAN, the ASONAM 2019 paper.
with open("nina_feminist","r") as Nina_feminist_file:
    for line in Nina_feminist_file.readlines():
        Nina_feminist_usernames.append(line.strip("\n"))

with open("nina_misogynist","r") as Nina_misogynist_file:
    for line in Nina_misogynist_file.readlines():
        Nina_msiogynist_usernames.append(line.strip("\n"))

# CESNA
with open("stanford_fem_hash_3","r") as stanford_feminist_file:
    for line in stanford_feminist_file.readlines():
        stanford_feminist_usernames.append(line.strip("\n"))

with open("stanford_mis_hash_3","r") as stanford_misogynist_file:
    for line in stanford_misogynist_file.readlines():
        stanford_msiogynist_usernames.append(line.strip("\n"))
```

```python
# How many tweets have I done
tweet_count = dict()
# What is the total times my tweet has been retweeted
retweet_count = dict()
# How many followers Do I have
follower_count = dict()
with open("gamergate.json", "r") as tweets_file:
    tweets = tweets_file.readlines()
    for line in tweets:
        author = ""
        rtauthor = ""
        followers = 0
        try:
            tweet = json.loads(line)
            try:
                author = tweet["user"]["screen_name"]
                rtauthor = tweet["retweeted_status"]["user"]["screen_name"]
                followers = int(tweet["user"]["followers_count"])
            except:
                try:
                    author = tweet["user"]["screen_name"]
                    followers = int(tweet["user"]["followers_count"])
                except:
                    continue
            # Count of my tweets: Direct or retweet
            if author in tweet_count:
                tweet_count[author] += 1
                # Count my followers
                follower_count[author] = followers
            else:
                tweet_count[author] = 1
                # Count my followers
                follower_count[author] = followers
            # Count of my retweets
            if rtauthor != "" :
                if rtauthor in retweet_count:
                    retweet_count[rtauthor] += 1
                else:
                    retweet_count[rtauthor] = 1
                
                if rtauthor in tweet_count:
                    tweet_count[rtauthor] += 1
                else:
                    tweet_count[rtauthor] = 1
        except:
            continue
```

```python
idx = 0
payload_objects = []
for user in tweet_count.keys():
    idx += 1
    Mturk_label = "N/A"
    Nina_label = "N/A"
    Network_label = "N/A"
    Stanford_label = "N/A"
    if(user in Mturk_feminist_usernames):
        Mturk_label = "Feminist"
    elif(user in Mturk_misogynist_usernames):
        Mturk_label = "Misogynist"
    if(user in Network_feminist_usernames):
        Network_label = "Feminist"
    if(user in Network_misogynist_usernames):
        Network_label = "Misogynist"
    if(user in Nina_feminist_usernames):
        Nina_label = "Feminist"
    if(user in Nina_msiogynist_usernames):
        Nina_label = "Misogynist"
    if(user in stanford_feminist_usernames):
        Stanford_label = "Feminist"
    if(user in stanford_msiogynist_usernames):
        Stanford_label = "Misogynist"
    rt_val = 0
    follower_val = -1
    if user in retweet_count:
        rt_val = retweet_count[user]
    if user in follower_count:
        follower_val = follower_count[user]

    payload_objects.append({ "id": idx, "uName": user, "mTurk": Mturk_label, "usc": Nina_label, "network": Network_label,
                "stanford": Stanford_label, "followers": follower_val, "totalTweets": tweet_count[user],
                "totalTimesRetweeted": rt_val })
```

```python
directory_path = "../data/"
if not os.path.exists(directory_path):
    os.makedirs(directory_path)
filepath = directory_path + "cleanData.json"
with open(filepath, "w+") as fwrite:
    json.dump(payload_objects, fwrite)
print("--- File Generated in %s seconds ---" % (time.time() - start_time))
```
